```python
df = df_copy
```

```python
def time_calc(row):
    calc = row["moving_time"]/row["elapsed_time"]
    return calc
```

```python
def elap_tot_sec(row):
    sec = row["elapsed_time"].total_seconds()
    return sec
```

```python
def mov_tot_sec(row):
    sec = row["moving_time"].total_seconds()
    return sec
```

```python
df
```

```python
df["elapsed_time_sec"] = df.apply(elap_tot_sec, 1)
df["moving_time_sec"] = df.apply(mov_tot_sec, 1)
df["time_ratio"] = df.apply(time_calc, 1)
df["calc"] = df["average_speed"] * df["time_ratio"]
df.loc[:, ["name", "average_speed", "time_ratio", "calc"]]
```
